```python
def filter_general(value):
    if '-' in str(value):
        return str(value).split('-')[1]
    elif '_' in str(value):
        return str(value).split('_')[0]
    else:
        return str(value)
```

```python
for i in ['age', 'annual_income', 'num_of_loan', 'outstanding_debt', 'monthly_balance']:
    data[i] = data[i].apply(filter_general)
    data[i] = data[i].astype(np.float64)
    print(i + 'Successfully Cleaned')
```

```python
data['change_credit_limit'] = data['changed_credit_limit'].apply(lambda x: x.split('-')[-1])
data.drop(data[data['changed_credit_limit'] =='_'].index, inplace=True)
data['changed_credit_limit'] = data['changed_credit_limit'].astype('float')
```

```python
data['occupation'] = data['occupation'].replace('_______', np.nan)
data['occupation'] = data['occupation'].fillna(np.random.choice(pd.Series(['Scientist', 'Teacher', 'Engineer', 'Entrepreneur','Developer', 'Lawyer', 'Media_Manager', 'Doctor', 'Journalist',
       'Manager', 'Accountant', 'Musician', 'Mechanic', 'Writer',
       'Architect'])))                                                
```

```python
data['credit_mix'] = data['credit_mix'].replace('_', np.nan)
data['credit_mix'] = data['credit_mix'].fillna(np.random.choice(pd.Series(['Standard', 'Good', 'Bad'])))
```

```python
data['payment_of_min_amount'] = data['payment_of_min_amount'].replace('NM', np.nan)
data['payment_of_min_amount'] = data['payment_of_min_amount'].fillna(np.random.choice(pd.Series(['Yes', 'No'])))
```
